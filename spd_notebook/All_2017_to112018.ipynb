```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("/Users/Stacy/Documents/WebDevelopment/code/spd/All_2017_to_11192018.csv")
data = pd.DataFrame(data)
```

```python
data.columns = ['cad_num', 'event_clearing', 'call_type', 'priority', 
                'initial_call_type', 'final_call_type', 'original_time_queued',
               'arrived_time', 'precinct', 'sector', 'beat']
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['original_time_queued'] = pd.to_datetime(data.original_time_queued)
```

```python
assault_calls = data[(data.final_call_type.str.contains('SEX')) | (data.final_call_type.str.contains('RAPE'))]
assault_calls
```

```python
assault_calls.dtypes
```

```python
assault_calls['final_call_type'].describe()
```

```python
crime_types = assault_calls[['final_call_type', 'arrived_time']]
crime_types.head()
```

```python
# Group by crime type
crime_type_group = crime_types.groupby('final_call_type')
crime_type_group.size()
```

```python
crime_type_group.count()
```

```python
my_plot = crime_type_group.count().plot(kind='bar', legend=None)
my_plot.set_xlabel("Crime Type")
my_plot.set_ylabel("Number of Calls")
```

```python
crime_types = assault_calls[['year','final_call_type', 'arrived_time']]
crime_types.count()
```

```python
crime_by_year_group = crime_types.groupby(['final_call_type', 'year']).count()
crime_by_year_group
```

```python
crime_by_year_group.unstack()
```

```python
my_plot2 = crime_by_year_group.unstack().plot(kind='bar',stacked=True,title='Crimes Reported per Year',figsize=(9,7))
my_plot2.set_xlabel("Crime Type")
my_plot2.set_ylabel("Number of Calls")
my_plot2.legend(['2017','2018'], loc=1,ncol=4)
```
